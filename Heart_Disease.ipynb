```python
x = dataset.drop(columns=['heart_disease', 'fbs', 'bmi', 'oldpeak'])
y = dataset['heart_disease']
```

```python
x.info()
```

```python
y.info()
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(
    x, y, random_state=42, test_size=0.2
)
```

```python
from sklearn.linear_model import LogisticRegression


lr = LogisticRegression(max_iter=1000)
lr.fit(xtrain, ytrain)

# Logical Correction: The problem is a classification task (heart_disease is 0 or 1),
# but the code uses regression models (...Regressor). This is fundamentally incorrect.
# The .score() method for regressors calculates the R-squared value, which is not
# suitable for classification and can give misleading negative scores.
# The correct approach is to use the classification counterparts of these models.

# Import the correct classifier models
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression # Changed from LinearRegression

# Use classifier models instead of regressors
models = [
    RandomForestClassifier(random_state=42),
    DecisionTreeClassifier(random_state=42),
    LogisticRegression(max_iter=1000, random_state=42) # Replaced LinearRegression
]

# Storing scores, so the dictionary is renamed for clarity
scores = {}

for model in models:
    model.fit(xtrain, ytrain)
    name = type(model).__name__
    # .score() for classifiers correctly returns accuracy
    score = model.score(xtest, ytest) * 100
    scores[name] = score

# Print the results to see the performance of each model
import pandas as pd
scores_df = pd.DataFrame(list(scores.items()), columns=['Model', 'Accuracy (%)'])
print(scores_df)

score = lr.score(xtest, ytest)
print(f"Model Accuracy: {score}")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report


predicted = lr.predict(xtest)

accuracy = accuracy_score(ytest, predicted)
print(f"Accuracy: {accuracy:.2f}")
print("-" * 30)

print("Confusion Matrix:")
print(confusion_matrix(ytest, predicted))
print("-" * 30)

print("Classification Report:")
print(classification_report(ytest, predicted))
```
